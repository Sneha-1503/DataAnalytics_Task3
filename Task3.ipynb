```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
df = pd.read_csv("products-100.csv")

print("Shape:", df.shape)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
print(df.isnull().sum())
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Category')
plt.title("Product Count by Category")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Availability')
plt.title("Product Availability Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Price'], bins=15, kde=True)
plt.title("Price Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Stock'], bins=15, kde=True)
plt.title("Stock Distribution")
plt.show()
```

```python
brand_counts = df['Brand'].value_counts()

brand_counts.plot(kind='bar', figsize=(8,5))
plt.title("Products per Brand")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Price', y='Stock')
plt.title("Price vs Stock")
plt.show()
```

```python
numeric_df = df[['Price','Stock']]

corr = numeric_df.corr()

print(corr)

sns.heatmap(corr,
            annot=True,
            cmap='coolwarm')

plt.title("Correlation Matrix")
plt.show()
```
